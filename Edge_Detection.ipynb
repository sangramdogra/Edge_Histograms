# Edge Histograms From Image

```python
import matplotlib.pyplot as plt
import cv2
import numpy as np
```

```python
#Loading square image
img= cv2.imread('square.jpg')
plt.figure(figsize = (5, 5))
plt.imshow(img)
plt.show()
```

```python
#Calculating derivatives from image
gx = cv2.Sobel(img, cv2.CV_32F, 1, 0)
gy = cv2.Sobel(img, cv2.CV_32F, 0, 1)
# Getting angles and magnitude
mag, ang = cv2.cartToPolar(gx, gy, angleInDegrees = True)
```

```python
#Horizontal derivatives
fig, axs = plt.subplots(figsize = (10,5), ncols = 2)
axs[0].imshow(gx)
axs[1].imshow(gy)
plt.show()
```

```python
plt.hist(ang.flatten(), bins = 10, density = True)
plt.show()
```
